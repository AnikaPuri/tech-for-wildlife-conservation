```
import math
import random 
import time

#-----------------------------------
# Adjascency List for Graph
class AdjNode:
    def __init__(self, value):
        self.vertex = value
        self.next = None

class Graph:
    def __init__(self, num):
        self.V = num
        self.graph = [None] * self.V

    # Add edges
    def add_edge(self, s, d):
        node = AdjNode(d)
        node.next = self.graph[s]
        self.graph[s] = node

        #node = AdjNode(s)
        #node.next = self.graph[d]
        #self.graph[d] = node

    # Print the graph
    def print_agraph(self):
        for i in range(self.V):
            print("Vertex " + str(i) + ":", end="")
            temp = self.graph[i]
            while temp:
                print(" -> {}".format(temp.vertex), end="")
                temp = temp.next
            print(" \n")

def is_cornerNode(node, side):
  if (node==0) or (node==side - 1) or (node==side*side - side) or (node==side*side - 1):
    return True

def is_sideNode(node, side):
  if (node%side == 0) or (node>0 and node<side) or node%side == side-1 or (node>side*side - side  and node<side*side - 1):
    return True

def is_centerNode(node,  side):
  if (is_sideNode(node) == False and is_cornerNode(node) == False):  
    return True


def generateGraph(side, graph):
  max_nodes = side*side

  for k in range(max_nodes):
    if is_cornerNode(k, side):  #corner
      if k == 0:  #upper-left
        graph.add_edge(k,side+1)
        graph.add_edge(k,side)
        graph.add_edge(k, 1)
      elif k == side - 1: #upper-right
        graph.add_edge(k,k+side)
        graph.add_edge(k,k+side-1)
        graph.add_edge(k,k-1)
      elif k == side*side - side: #lower-left
        graph.add_edge(k,k+1)
        graph.add_edge(k,k-side+1)
        graph.add_edge(k,k-side)
      elif k == side*side - 1: #lower-right
        graph.add_edge(k,k-1)
        graph.add_edge(k,k-side)
        graph.add_edge(k,k-side-1)

    elif is_sideNode(k, side): #side
      if (k%side == 0): #left-column
        graph.add_edge(k,k+side+1)
        graph.add_edge(k,k+side)
        graph.add_edge(k,k+1)
        graph.add_edge(k,k-side+1)
        graph.add_edge(k,k-side)
      elif (k%side == side-1): #right-column
        graph.add_edge(k,k+side)
        graph.add_edge(k,k+side-1)
        graph.add_edge(k,k-1)
        graph.add_edge(k,k-side)
        graph.add_edge(k,k-side-1)
      elif (k>0 and k<side): #top-row
        graph.add_edge(k,k+side+1)
        graph.add_edge(k,k+side)
        graph.add_edge(k,k+side-1)
        graph.add_edge(k,k+1)
        graph.add_edge(k,k-1)
      elif (k>side*side - side and k<side*side - 1): #bottom-row
        graph.add_edge(k,k+1)
        graph.add_edge(k,k-1)
        graph.add_edge(k,k-side+1)
        graph.add_edge(k,k-side)
        graph.add_edge(k,k-side-1)

    else: #center
      graph.add_edge(k,k+side+1)
      graph.add_edge(k,k+side)
      graph.add_edge(k,k+side-1)
      graph.add_edge(k,k+1)
      graph.add_edge(k,k-1)
      graph.add_edge(k,k-side+1)
      graph.add_edge(k,k-side)
      graph.add_edge(k,k-side-1)

  return graph    
```

```
#-----------------
#INPUTS
grid_size = 200
max_animals = 9
begin_node = 0
end_node = 0
path_length = 10
#-----------------

# Create graph and edges
max_nodes = grid_size*grid_size
graph = Graph(max_nodes)
graph = generateGraph(grid_size, graph)
#graph.print_agraph()

actual_cost = [0] * max_nodes
for i in range(max_nodes):
  actual_cost[i] = random.randint(0,max_animals)
print('Animal Distribution:', actual_cost)
```

```
# 
# Return all distinct simple paths from "originNode" to "targetNode". 
# We are given the graph in the form of a adjacency list "nodeToNodes". 
# 
def getAllSimplePaths(originNode, targetNode, graph):
  dronePath = helpGetAllSimplePaths(str(targetNode), 
                                 [str(originNode)], 
                                 set(), 
                                 graph, 
                                 list()) 
  return dronePath


# 
# Return all distinct simple paths ending at "targetNode", continuing 
# from "currentPath". "usedNodes" is useful so we can quickly skip 
# nodes we have already added to "currentPath". When a new solution path 
# is found, append it to "answerPaths" and return it. 
#     
def helpGetAllSimplePaths(targetNode, 
                          currentPath, 
                          usedNodes, 
                          graph, 
                          answerPath):           
    global count, total_count, new_cost, min_cost, path_cost, distance, path_length, startPathTime, endPathTime, timed_out, node_in_region

    endPathTime = time.time()

    if (endPathTime - startPathTime) > 1000: #will exit if a more efficienit path is not found in more than 1000 seconds
      #print(endPathTime, startPathTime)
      timed_out = True
      return answerPath

    #[-1] index means last element  in the Path
    currentNode = currentPath[-1] 

    #This ensures we are in permitted region only which is constrainted by path_length
    if currentNode:
      if not node_in_region[int(currentNode)]:
        return answerPath

    #print(currentPath)
    if (currentNode == targetNode) and (distance > 1): 
        total_count = total_count+1
        #print('total_count-', total_count, currentPath, path_cost)
        #append the answer path only if the path length is same as desired
        if len(list(currentPath)) == path_length:
          count = count + 1
          #print(count, 'cost:', path_cost, currentPath)
          if path_cost < min_cost :
            min_cost = path_cost;
            startPathTime = time.time()
            print('Min cost = ', min_cost, 'Path: ', currentPath)
            answerPath.clear()
            answerPath.append(list(currentPath))

    else: 
        temp = graph.graph[int(currentNode)]
        '''
        print(currentNode)
        t = temp
        while t:
            print(" -> {}".format(t.vertex), end="")
            t = t.next
        print(" \n")
        '''
        while temp: 
            neighbor = str(temp.vertex)
            if neighbor not in usedNodes: 
                cost  = new_cost[int(neighbor)]             
                currentPath.append(neighbor) 
                path_cost = path_cost + cost
                usedNodes.add(neighbor) 
                distance = distance + 1
                #Explore further only if path cost so far is less than min_cost 
                # and distance from original is less than path_length constraint
                if (path_cost <= min_cost) and (distance <= path_length) :
                    helpGetAllSimplePaths(targetNode, 
                                          currentPath, 
                                          usedNodes, 
                                          graph, 
                                          answerPath)
                
                path_cost = path_cost - cost
                usedNodes.remove(neighbor) 
                distance = distance - 1
                currentPath.pop() 
            temp = temp.next
            
    return answerPath


def in_region(node, path_length, max_nodes, grid_size):
  if (path_length/2 >= grid_size):
    return True
  else:
    new_grid_size = int(path_length/2 + 1)
    for y in range(new_grid_size):
      if (node%grid_size == y):
        return True 
    return False


#-----------------
max_value = 0
for i in range(max_nodes):
  if actual_cost[i] > max_value: 
    max_value = actual_cost[i]

new_cost = [0]*max_nodes
for i in range(max_nodes): 
  new_cost[i] = max_value - actual_cost[i]

min_cost = max_value * path_length

count = total_count = 0
path_cost = new_cost[begin_node] 
distance = 1

node_in_region = [False] * max_nodes
for i in range(max_nodes):
  if in_region(i,path_length, max_nodes, grid_size) == True:
    node_in_region[i] = True
  else:
    node_in_region[i] = False

#print(node_in_region)

timed_out = False
startPathTime = time.time()
start = time.time()
dronePath = getAllSimplePaths(begin_node, end_node, graph)
end = time.time()

if timed_out == True:
  print("Path Search timed out after 15 minutes of searching for a better solution than min cost of: ", min_cost)

print("Time: ", end - start)

print('Total Paths explored  = ', str(total_count))
print('Number of paths of length ', str(path_length), ' = ', str(count))

max_cost = max_value*path_length - min_cost
print('Optimal Drone Path Cost = ', max_cost)
print('Drone Path = ', dronePath)

```
