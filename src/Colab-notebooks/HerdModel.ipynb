```python
import os
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

max_count = 30 #columnn
result = [[0 for x in range(max_count)] for y in range(9)]

rows = 7000
columns = 9 #species

#arr is an array that stores info for a given CSV file
#  rows are for frame_id 
#  columns is species label, i.e., 0 to 8
arr = [[0 for x in range(columns)] for y in range(rows)]

basepath = '/content/drive/My Drive/Real_annotation'
max = 0
for entry in os.listdir(basepath):
    file_name = os.path.join(basepath, entry)
    if(entry != 'out.csv') :
      if os.path.isfile(file_name):
          print(file_name)
          
          #intialize arr to 0
          arr = [[0 for x in range(columns)] for y in range(rows)]
          df = pd.read_csv(file_name)

          for ind in df.index:
            species = df.iloc[ind,7]
            if(species != -1):
              frame_index = df.iloc[ind,0]
              arr[frame_index][species] = arr[frame_index][species] + 1

          for species in range(9) :
            for  frame_index in range(7000) :
              count = arr[frame_index][species]
              if(count > max):
                max = count
              if(count > 0) :
                result[species][count] = result[species][count] +  1

print(max)
print(result)

import csv

with open("/content/drive/My Drive/Real_annotation/out.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(result)




        
```
